```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sb 
import plotly.express as px
import plotly.graph_objects as go
from scipy.io import arff
```

```python
balance_data = arff.loadarff('covidbalance.arff')
balancedata = pd.DataFrame(balance_data[0])
balancedata.info()
```

```python
balancedata['Patient addmited to regular ward (1=yes, 0=no)'].replace({
    0.305052:0, 0.964105:1, 0.317157:0, 0.789417:1,0.374353:0, 0.298387:0, 0.972338:1, 0.033762:0, 0.027774:0, 0.781857:1,0.513901:1, 0.42999:0 , 0.294143:0, 0.947679:1, 0.357703:0, 0.156971:0,0.715578:1, 0.555518:1
    }, inplace=True)
balancedata['Patient addmited to semi-intensive unit (1=yes, 0=no)'].replace({
    0.327156:0, 0.546385:1, 0.654931:1, 0.409158:0,0.026229:0, 0.067758:0, 0.620755:1, 0.71071:1 , 0.644962:1, 0.469032:0,0.947893:1, 0.549252:0
    }, inplace=True)
balancedata['Patient addmited to intensive care unit (1=yes, 0=no)'].replace({
    0.342549:0, 0.786461:1, 0.032755:0, 0.081505:0,0.493074:0, 0.460656:0, 0.89549:1 , 0.158321:0, 0.148163:0
    }, inplace=True)
```

```python
#heat map
# Correlation
corr = balancedata.corr()
# Covariance
cov = balancedata.cov() 
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(15, 8))
# Generate a custom diverging colormap
cmap = sb.diverging_palette(10, 10, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sb.heatmap(corr, annot=True, linewidths=.5, cmap="YlGnBu")
plt.show()
```
