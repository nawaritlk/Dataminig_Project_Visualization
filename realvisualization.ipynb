```python
dataset = balancedata.groupby('SARS-Cov-2 exam result')[whiteBloodCell].mean()
dataset
```

```python
indx = np.arange(len(whiteBloodCell))
print(dataset.T)

Lymphocytes_means_negative = list(dataset.T['be_negative'])
Lymphocytes_means_positve = list(dataset.T['be_positive'])
bar_width = 0.35
fig, ax = plt.subplots()
barLymphocytes_negative = ax.bar(indx - bar_width/2,Lymphocytes_means_negative,bar_width,Label ='be_negative')
barLymphocytes_positive = ax.bar(indx + bar_width/2,Lymphocytes_means_positve,bar_width,Label ='be_posivtive')

ax.set_xticks(indx)
ax.set_xticklabels(whiteBloodCell)

#ax.set_yticks()
#ax.set_yticklabels(Patient)

#ax.legend()

for i in indx:
    ax.annotate('{0:.0f}'.format(barLymphocytes_negative[i].get_height()),
                xy=(barLymphocytes_negative[i].get_x() + barLymphocytes_negative[i].get_width() /2, barLymphocytes_negative[i].get_height()),
                xytext=(0,5),
                textcoords='offset points',
                ha='center',
                va='bottom'
               )
    
for i in indx:
    ax.annotate('{0:.0f}'.format(barLymphocytes_positive[i].get_height()),
                xy=(barLymphocytes_positive[i].get_x() + barLymphocytes_positive[i].get_width() /2, barLymphocytes_positive[i].get_height()),
                xytext=(0,5),
                textcoords='offset points',
                ha='center',
                va='bottom'
               )
plt.show()
```
